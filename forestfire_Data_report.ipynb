```python
df=pd.read_csv('forestfires.csv')
sales_report = pd.pivot_table(df, index=["month", "day"], values=["temp","wind","rain"],
                           aggfunc=[np.sum, np.mean,np.std], fill_value=0)
print (sales_report)
```

```python
new=(sales_report.groupby('month'))
newmean=new.mean()
new=new.std()
# f=plt.figure(figsize=(100,100))
axmean=newmean[['mean']].plot(kind='bar',title="ForestFire Annual Report (Mean)",figsize=(15,10),legend=True,fontsize=12)
ax=new[['std']].plot(kind='bar',title="ForestFire Annual Report(Standard Deviation)",figsize=(15,10),legend=True,fontsize=12)
ax.set_xlabel("Month",fontsize=12)
ax.set_ylabel("Data")
axmean.set_xlabel("Month",fontsize=12)
axmean.set_ylabel("Data")
plt.show()
new
plt.savefig("deepanshi.pdf")
```
